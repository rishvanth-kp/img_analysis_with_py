## Converting a grayscale image to a binary image

Binary thresholding is the process of converting a grayscale image into a binary image. 
This conversion is done by picking a specic threshold value that is in the range of the pixel intensityes, i.e between 0 and 64k for a 16-bit image. All the pixles in the image that have an intesntsy value below this threhshols are set to 0 and the intensity values above this threshold are set to 1. 

To binary threshold can be accomplished with the comparivete operator `<` or `>`, which returns a boolean values pixels meeting the condisiton. For exampale:

```python
# Load the required libraries
import skimage as ski
import numpy as np
import matplotlib.pyplot as plt

# Load the coins data
img = ski.data.coins()

img_thresh = img < 100


fig, ax =  plt.subplots(1, 2)
ax[0].imshow(img, cmap = 'gray')
ax[1].imshow(img_thresh, cmap = 'gray')
plt.axis('off')
```

## Image intensity histogram

An image intensity historam is an histogram of the number of pixles of at each intensieyt values. The x-axis is the pixel values, and hace in the range newttne 0 and the 64k for a 16 bit image. The y-axis is the number of pixles at a particaurl internetery vaoue.

We can compute the intensity histogram using the `xyz` fumction: 

```python
hist, hist_centers = ski.exposure.histogram(img)

fig, ax =  plt.subplots(1, 2, figsize=(6,3))
ax[0].imshow(img, cmap = 'gray')
ax[1].plot(hist_centers, hist)
```

The histogram tells us a lot about the image. For example, from the histgram above.. 

A "darker" image has a histogram peak at the left end whereas a "brigther" image has pixles on the right end of the histrgarm. An image that uses the entiture range of the hgistgram would enable the best differntatob between distucnt image intensituytes. 

An grayshcale image that has a distinct foreground and backgournd pixels have a bimodal histogram. A peak on the left of the histogram would correspond to the backgound and the peak on the right to the foregound. 

```python
img = ski.data.coins()
img = ski.exposure.equalize_hist(img, 2)

hist, hist_centers = ski.exposure.histogram(img)


fig, ax =  plt.subplots(1, 2, figsize=(6,3))
ax[0].imshow(img, cmap = 'gray')
ax[1].plot(hist_centers, hist)
```
